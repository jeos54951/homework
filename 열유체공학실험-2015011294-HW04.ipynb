```
import math
import numpy as np
from numba import jit,vectorize, cuda
import os
os.environ['NUMBAPRO_LIBDEVICE'] = "/usr/local/cuda-10.0/nvvm/libdevice"
os.environ['NUMBAPRO_NVVM'] = "/usr/local/cuda-10.0/nvvm/lib64/libnvvm.so"
a= int(input("input a:"))
def function(a):
  for b in range(1,10):
    print(a,"x",b,"=",a*b,"입니다.")

  return a
      
  
print('NORMAL')
function(int(a))
print()
  


def gama(a):
  x1 = np.arange(1,10)
  x2 = x1*a
  x1 = x2.reshape((9,1))

  return x1

print('NUMPY')
print(gama(a))
print()

@jit(nopython=True)
def alpha(a):
  x3 = np.arange(1,10)
  x4 = x3*a
  x3 = x4.reshape((9,1))
  print(x3)
  return x3
print('NUMBA JIT')
alpha(int(a))
print()


@vectorize(['int64(int64,int64)'],target='cuda')
def beta(y,z):
  return y*z


x7 = np.arange(1,10)
D = beta(a,x7).reshape((9,1))

print('CUDA')
print(D)
print()

print("Normal Compiling Time(s) : ", end = "")
%timeit function

print("Numpy Compiling Time(s) : ", end = "")
%timeit gama

print("Numba Jit Compiling Time(s) : ", end = "")
%timeit alpha

print("Cuda Compiling Time(s) : ", end = "")
%timeit beta
```
